```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Use case :1. Create a  geographic  map of states with low literacy rates.
```

```python
df = pd.read_csv("india-districts-census-2011.csv")
df
```

```python
cen = df[["State name","Population","Literate"]]
cen
```

```python
grouped = cen.groupby('State name')
cen_sum = grouped.sum()
cen_sum
```

```python
cen_lit = (cen_sum["Literate"] / cen_sum["Population"]*100)
cen_lit
```

```python
cen_sort = cen_lit.sort_values(axis=0, ascending=True, inplace=False, kind='quicksort')
cen_sort.head(10)
```

```python
#bar graph for the fist 10 low literacyrate states..
data = cen_sort.head(10).plot.bar(title = "Low literacy rates",color="g")
data.set_xlabel("State Names")
data.set_ylabel("Literacy Rates(%)")
plt.show()
```

```python
#Use case: 2. Find out most similar districts in Bihar and Tamil Nadu. Similarity can be based on any of
#the columns from the data.
```

```python
df1 = df[(df["State name"] == "BIHAR")]
df2 = df[(df["State name"] == "TAMIL NADU")]
df3 = df1[["State name","District name","Literate"]]
df3
```

```python
df4 = df2[["State name","District name","Literate"]]
df4
```

```python
for i, s, d, l in df3.itertuples():
    print(d)
    print(l)
```

```python
for i, s, d, l in df4.itertuples():
    print(d)
    print(l)
```

```python
d = 0.005
df7 = pd.DataFrame({'District': [np.nan],'Literate': [np.nan]})
df8 = pd.DataFrame({'District': [np.nan],'Literate': [np.nan]})
for bi, bs, bd, bl in df3.itertuples():
    for ti, ts, td, tl in df4.itertuples():
        if(bl-d*bl<tl and tl<bl+d*bl):
            temp1 = pd.DataFrame({'District': [bd],'Literate': [bl]})
            temp2 = pd.DataFrame({'District': [td],'Literate': [tl]})
            frames = [df7, temp1]
            df7 = pd.concat(frames)
            df7 = df7.dropna()
            frames = [df8, temp2]
            df8 = pd.concat(frames)
            df8 = df8.dropna()

df7.count()
```

```python
#Use Case:3.How does the mobile penetration vary in regions (districts or states) with high or low
#agricultural  workers
```
